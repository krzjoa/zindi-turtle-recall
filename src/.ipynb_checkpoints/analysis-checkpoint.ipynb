## Turtle Recall (Zindi challenge)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
import cv2 as cv
import PIL
```

```python
# Loading data
TRAIN_CSV_PATH = "../data/train.csv"
TEST_CSV_PATH = "../data/test.csv"
EXTRA_IMAGES_PATH = "../data/extra_images.csv"
SAMPLE_SUBMISSION_PATH = "../data/sample_submission.csv"
IMAGES_PATH = "../data/images"

# CSV files
train_csv = pd.read_csv(TRAIN_CSV_PATH)
test_csv  = pd.read_csv(TEST_CSV_PATH)
extra_images = pd.read_csv(EXTRA_IMAGES_PATH)
sample_submission = pd.read_csv(SAMPLE_SUBMISSION_PATH)

all_imgs = pd.concat([train_csv, extra_images, test_csv])


# List of images
imgs = os.listdir(IMAGES_PATH, )
imgs = [os.path.join(IMAGES_PATH, x) for x in imgs]
```

```python
all_imgs.head()
```

```python
## Loading an image
sample_image = cv.imread(imgs[0])
sample_image.shape

plt.imshow(sample_image)
plt.title("TURTLE")
plt.show()
```

```python
def show_sample_image(idx: int):
    # Img path 
    img_path = imgs[idx]
    img_id = os.path.basename(img_path)
    img_id = os.path.splitext(img_id)[0]
    
    row = all_imgs[all_imgs['image_id'] == img_id]
    row_as_title = row.to_csv(header=None, index=False).strip('\n').split('\n')
    
    # Loading an image
    image = cv.imread(img_path)

    plt.imshow(image)
    plt.title(row_as_title)
    plt.show()
    
    print(img_id)
    

def get_filename(x: str)-> str:
    img_id = os.path.basename(x)
    img_id = os.path.splitext(img_id)[0]
    return img_id  
```

```python
## Possible sizes
image_sizes = [
    PIL.Image.open(x).size for x in imgs
]
```

```python
image_sizes = pd.DataFrame({
    'image_path': imgs,
    'image_width': [x[0] for x in image_sizes],
    'image_heigth': [x[1] for x in image_sizes]
})
```

```python
image_sizes['image_path'] = image_sizes.image_path.apply(get_filename)
```

```python
# Plot typical size of images
plt.plot(image_sizes.image_width, image_sizes.image_heigth, 'o')
plt.show()
```
